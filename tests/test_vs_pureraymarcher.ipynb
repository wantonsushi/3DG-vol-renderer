### Testing Pure-Raymarcher

```python
import mitsuba as mi
mi.set_variant('scalar_rgb')

import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
ref_img = imageio.imread('../build/sphere_ref.ppm').astype(np.float32) / 255.0
user_img = imageio.imread('../build/sphere_test.ppm').astype(np.float32) / 255.0

diff = user_img - ref_img

signed_error = np.mean(diff, axis=2)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(user_img)
axs[0].set_title('Raymarcher (Analytical Transmittance)')
axs[0].axis('off')

axs[1].imshow(ref_img)
axs[1].set_title('Raymarcher (Raymarching Transmittance)')
axs[1].axis('off')

im = axs[2].imshow(signed_error, cmap='seismic',
                   vmin=-np.max(np.abs(signed_error)),
                   vmax=np.max(np.abs(signed_error)))
axs[2].set_title('Signed Error')
axs[2].axis('off')
fig.colorbar(im, ax=axs[2], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
